```python
def plan_trip(destination, days=3, budget="budget-friendly", travel_type="solo"):
    prompt = (
        f"Plan a {days}-day {budget} {travel_type} trip to {destination}. "
        "Include transportation, accommodation, daily activities, and estimated cost. "
        "Return the output day-wise as a nicely formatted itinerary. "
        "Use Indian Rupees (INR ₹) for all costs."
    )
    response = model.generate(prompt=prompt)
    return response['results'][0]['generated_text']
print(plan_trip("Delhi", days=4, budget="luxury", travel_type="family"))
```
